```python
store_data=pd.read_csv("D://datasets//store_data.csv",header=None)
```

```python
store_data.fillna(0,inplace=True)
```

```python
store_data.head()
```

```python
list_store=[]
for i in range(0,len(store_data)):
    list_store.append([str(store_data.values[i,j]) for j in range(0,20)])
    
```

```python
rules=apriori(list_store,min_support=0.003,min_confidence=0.2,min_lift=2,min_length=2)
rules
```

```python
results=list(rules)
results
```
