```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np

x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2))
```

```python
pylab.plot(x,y,'r')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g--')
```

```python
from matplotlib import pyplot as plt
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
    
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig=plt.figure(figsize=(8,8),dpi=50)

fig.add_subplot()

plt.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(figsize=(8,8),dpi=50)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title("title")
axes.plot(x,y,'r')
```

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
x,y=np.meshgrid(x,y)
z1=np.exp(-x**2-y**2)
z2=np.exp(-(x-1)**2-(y-1)**2)
z=(z1-z2)*2
print(x)
print(y)
```

```python
fig,ax=plt.subplots()
cs=ax.contour(x,y,z)
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title('contout map')
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig=plt.figure(figsize=(14,6))
ax=fig.add_subplot(1,2,1,projection='3d')#always there should be 1 nly(atlast)
ax.plot_surface(x,y,z,rstride=4,cstride=4,linewidth=5)
```
